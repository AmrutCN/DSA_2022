```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv (r'C:\Users\ADAM\Downloads\Advertising.csv')
df = pd.DataFrame(data, columns= ["TV","Radio","Newspaper","Sales"])
print(df)
```

```python
df.describe()
```

```python
df.sum()
```

```python
df_tv = df[(df['TV']> df['Radio']) & (df['TV']> df['Newspaper'])]
df_Radio = df[(df['Radio']> df['TV']) & (df['Radio']> df['Newspaper'])]
df_Newspaper = df[(df['Newspaper']> df['Radio']) & (df['Newspaper']> df['TV'])]
print("no of records with highest tv ad cost",len(df_tv.index),"\nno of records with highest radio ad cost",len(df_Radio.index),"\nno of records with highest newspaper ad cost",len(df_Newspaper.index))
```

```python
total_ad_mean=[df['TV'].mean(),df['Radio'].mean(),df['Newspaper'].mean()]
avg_sale_mean=[df_tv['Sales'].mean(),df_Radio['Sales'].mean(),df_Newspaper['Sales'].mean()]
avg_ad_mean=[df_tv['TV'].mean(),df_Radio['Radio'].mean(),df_Newspaper['Newspaper'].mean()]
```

```python
total=['TV','Radio','Newspaper']
plt.axhline(y=df['Sales'].mean(), color='r', linestyle='--',label="total sale mean")
plt.plot(total,total_ad_mean,color='Blue',label="total ad cost mean")
plt.plot(total,avg_ad_mean,color='green',label="mean of ad cost in respective category ")
plt.plot(total,avg_sale_mean,color='yellow',label="mean of sales in respective category ")
plt.rcParams['figure.figsize'] = [20, 20]
plt.legend()
plt.show
```
