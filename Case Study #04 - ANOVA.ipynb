```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv(r'C:\Users\ADAM\Downloads\Sales_data_zone_wise.csv')
df= pd.DataFrame(data, columns= ['Month','Zone - A','Zone - B','Zone - C','Zone - D'])
df
```

```python
df.describe()
```

1.The mean sales generated by each zone.

```python
zonesale_mean = [df['Zone - A'].mean(),df['Zone - B'].mean(),df['Zone - C'].mean(),df['Zone - D'].mean()]
total=['Zone - A','Zone - B','Zone - C','Zone - D']
plt.plot(total,zonesale_mean,color='Blue')
n=0
for i in total:
    plt.text(i,zonesale_mean[n],int(zonesale_mean[n]), size=20)
    n=n+1
# plt.plot(total,dfa,color='Red')
plt.title('Zone-Wise Sale Mean')
plt.rcParams['figure.figsize'] = [20, 20]
plt.show()
```

2.Total sales generated by all the zones for each month

```python
df['Total Monthly Sale']=df['Zone - A']+df['Zone - B']+df['Zone - C']+df['Zone - D']
df
```

```python
plt.barh(df['Month'],df['Total Monthly Sale'])
plt.title('Monthly Total Sale')
plt.rcParams['figure.figsize'] = [20, 20]
plt.grid()
plt.show
```

```python
print("Each zones generates the different amount of sales")
# df.loc['29'] = ['Total Count',df['Zone - A'].sum(),df['Zone - B'].sum(),df['Zone - C'].sum(),df['Zone - D'].sum(),df['Total Monthly Sale'].sum()]
dft = [df['Zone - A'].sum(),df['Zone - B'].sum(),df['Zone - C'].sum(),df['Zone - D'].sum()]
# dfa = [int(x / 28) for x in dft]
dft
```
