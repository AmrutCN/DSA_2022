5. Check and handle outliers in at least 3 columns in the dataset

```python
x=df[df['Pclass'] == 2]['Fare']
plt.boxplot(x)
plt.show()
```

```python
q1 = np.percentile(x,25,interpolation='midpoint')
q2 = np.percentile(x,50,interpolation='midpoint')
q3 = np.percentile(x,75,interpolation='midpoint')
print(q1,q2,q3)
Iqr = q3-q1
l_lim=q1-1.5*Iqr
u_lim=q3+1.5*Iqr
print(Iqr,l_lim,u_lim)
outlier=[]
for X in x:
    if((X>u_lim)|(X<l_lim)):
        outlier.append(X)
print(outlier)
print(len(outlier))
```

```python
print(df[(df['Pclass'] == 2)&(df['Fare']>u_lim)].index)
```
