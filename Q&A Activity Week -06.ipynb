```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_excel(r'C:\Users\ADAM\Downloads\Dataset_w6.xlsx')
df= pd.DataFrame(data, columns= ['country','Region','Surface area (km2)','Population in thousands (2017)','GDP per capita (current US$)','Agricultural production index ','International trade: Imports (million US$)'])
print(df)
```

2. Examine the data by using visualizations

```python
plt.plot(df['country'],df['International trade: Imports (million US$)'],color='Blue')
# plt.bar(df['country'],df['Agricultural production index '],color='Green')
plt.scatter(df['country'],df['GDP per capita (current US$)'],color='Yellow')
plt.bar(df['country'],df['Population in thousands (2017)'],color='Red')
# plt.plot(df['country'],df['Surface area (km2)'],color='orange')
plt.xticks(color='w')
plt.rcParams['figure.figsize'] = [20, 20]
plt.show
```

3. Find out how many countries with respect to the region are available
in the dataset

```python
df.Region.unique()
```

```python
regions=df.Region.unique()
region_dict={}
for region in regions:
    count=df[df['Region'] == region]
    region_dict.update({region:len(count)})
region_dict
```

```python
plt.barh(regions,region_dict.values())
plt.rcParams['figure.figsize'] = [20, 20]
plt.show
```

4a. On average which region has the highest GDP per capita (current US$)

```python
gdppc_dict={}
for region in regions:
    count=df[df['Region'] == region]
    gdppc_dict.update({region:count['GDP per capita (current US$)'].mean()})
gdppc_dict                
```

```python
plt.barh(regions,gdppc_dict.values())
plt.rcParams['figure.figsize'] = [20, 20]
plt.show
```

4b. On average which region has the highest International trade: Imports (million US$)

```python
iti_dict={}
for region in regions:
    count=df[df['Region'] == region]
    iti_dict.update({region:count['International trade: Imports (million US$)'].mean()})
iti_dict 
```

```python
plt.barh(regions,iti_dict.values())
plt.rcParams['figure.figsize'] = [20, 20]
plt.show
```

```python
plt.plot(gdppc_dict.keys(),gdppc_dict.values(),color='Blue')
plt.axhline(y=df['GDP per capita (current US$)'].mean(), color='r', linestyle='--')
for i in gdppc_dict:
    plt.text(i,gdppc_dict[i],i, size=12,rotation=45)
plt.rcParams['figure.figsize'] = [20, 20]
plt.xticks(color='w')
plt.grid()
plt.show
```
